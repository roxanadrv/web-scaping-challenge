## Deliverable 2: Scrape and Analyze Mars Weather Data

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html=browser.html
weather_soup=soup(html,'html.parser')
```

```python
# Extract all rows of data
table_rows=weather_soup.find_all('tr')
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
weather_data=[]

# Loop through the scraped data to create a list of rows
for row in table_rows[1:]:
    cols=row.find_all('td')
    cols=[col.text.strip() for col in cols]
    weather_data.append(cols)
    

    
# Display the weather_data list
for row in weather_data:
    print(row)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
    
header_row = table_rows[0]
headers = [header.get_text().strip() for header in header_row.find_all('th')]


df=pd.DataFrame(weather_data, columns=headers)
```

```python
# Confirm DataFrame was created successfully
df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
print(df.dtypes)
```

```python
# Change data types for data analysis
df['id']=df['id'].astype(int)
df['terrestrial_date']=pd.to_datetime(df['terrestrial_date'])
df['sol'] = df['sol'].astype(int)
df['ls'] = df['ls'].astype(int)
df['month'] = df['month'].astype(int)
df['min_temp'] = df['min_temp'].astype(float)
df['pressure'] = df['pressure'].astype(float)
```

```python
# Confirm type changes were successful by examining data types again
print(df.dtypes)
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
number_of_months_on_mars=df['month'].nunique()
print(f'Number of Martian months in the dataset: {number_of_months_on_mars}')
```

```python
# 2. How many Martian days' worth of data are there?
number_of_martian_days = df['sol'].nunique()
print(f'Number of Martian days in the dataset: {number_of_martian_days}')
```

```python
# 3. What is the average low temperature by month?
average_temp_by_month = df.groupby('month')['min_temp'].mean()
print(f'Average low temperature by month is: {average_temp_by_month}')
```

```python
# Plot the average temperature by month
average_temp_by_month.plot(kind='bar',figsize=(12,7),title='Average Low Temperature by Martian Month',xlabel='month',ylabel='temperature in Celsius')
plt.show()
```

```python
# Calculate the average temperature by month.
average_temp_by_month = df.groupby('month')['min_temp'].mean()
# Sort the months by average temperature.
sorted_months_by_temp = average_temp_by_month.sort_values()

# Plot the sorted data.
sorted_months_by_temp.plot(kind='bar', figsize=(12, 7), title='Average Temperature in Celsius by Martian Month')
# Set the x and y labels.
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius')
# The x-axis will display the month numbers sorted from the coldest to the warmest month.
plt.xticks(range(len(sorted_months_by_temp)), [f' {month}' for month in sorted_months_by_temp.index])

# Identify the coldest and hottest months in Curiosity's location
coldest_month = average_temp_by_month.idxmin()
warmest_month = average_temp_by_month.idxmax()
print(f'Coldest Martian month: {coldest_month}')
print(f'Warmest Martian month: {warmest_month}')

plt.show()
```

```python
# 4. Average pressure by Martian month
average_pressure_by_month = df.groupby('month')['pressure'].mean()
print(f'Average pressure by month is: {average_pressure_by_month}')
```

```python

# Identify months with the lowest and highest average atmospheric pressure
lowest_pressure_month = average_pressure_by_month.idxmin()
highest_pressure_month = average_pressure_by_month.idxmax()
print(f'Month with the lowest average pressure: {lowest_pressure_month}')
print(f'Month with the highest average pressure: {highest_pressure_month}')

# Plot the average pressure by month
average_pressure_by_month.plot(kind='bar',figsize=(12,7), title='Average Pressure by Martian Month',xlabel='month',ylabel='Atmospheric Pressure')
plt.show()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?


# To visualize the estimation of the Martian year based on minimum temperatures:
df.plot(figsize=(12,7),x='terrestrial_date', y='min_temp', title='Martian Daily Minimum Temperature Over Time')
plt.xlabel('Terrestrial Date')
plt.ylabel('Minimum Temperature (Â°C)')
plt.show()

date_difference = (df['terrestrial_date'].max() - df['terrestrial_date'].min()).days
print(f'Estimated number of Earth days in the dataset span: {date_difference}')

# Assuming three full Martian cycles are observed in the plot
number_of_martian_years = 3
martian_year_in_earth_days = date_difference / number_of_martian_years
print(f'Three Martian years (full cycle) can be seen from the plot within the {date_difference} Earth days span in the dataset.')
print(f'Then a Martian year is approximately {date_difference}/{number_of_martian_years}={martian_year_in_earth_days:.2f} Earth days based on the dataset.')


# The commonly accepted value for a Martian year in Earth days
accepted_martian_year_in_earth_days = 687
print(f'A Martian year is equivalent to {accepted_martian_year_in_earth_days} Earth days according to scientific consensus.')
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
